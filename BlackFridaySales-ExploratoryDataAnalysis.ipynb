### Analyzing Product Category with Purchase

```python
def format_thousands(x, _):
    if x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    elif x >= 1_000:
        return f'{x/1_000:.0f}K'
    return f'{x:.0f}'

formatter = FuncFormatter(format_thousands)

top_category = df.groupby('Product_Category_1')['Purchase'].sum().sort_values(ascending = False).head(5)
bottom_category = df.groupby('Product_Category_1')['Purchase'].sum().sort_values(ascending = False).tail(5)

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Top 5 Product Categories
top_category.plot(kind='bar', color='#4CAF50', ax=axs[0])
axs[0].set_title('Top 5 Product Categories by Purchase Amount', fontsize=12, weight='bold')
axs[0].set_xlabel('Product Category')
axs[0].set_ylabel('Total Purchase Amount')
axs[0].tick_params(axis='x', rotation=0)
axs[0].yaxis.set_major_formatter(formatter)

# Bottom 5 Product Categories
bottom_category.plot(kind='bar', color='#FF7043', ax=axs[1])
axs[1].set_title('Bottom 5 Product Categories by Purchase Amount', fontsize=12, weight='bold')
axs[1].set_xlabel('Product Category')
axs[1].set_ylabel('Total Purchase Amount')
axs[1].tick_params(axis='x', rotation=0)
axs[1].yaxis.set_major_formatter(formatter)

plot.tight_layout()
```

### Insights:
- Products from **category 1** contributed the most towards revenue while the products from **category 13** contributed the least to the revenue.

- Marketers can come up with strategies to increase the sales of products in the bottom five categories.

```python
sns.lmplot(x='Stay_Years_Num', y='Purchase', data=user_purchase, scatter_kws={'alpha':0.4})
plot.title('Stay Duration vs Total Spending')
plot.xlabel('Stay Years')
```

### Insights:
- No strong correlation was observed as distribution of total purchases appears relatively uniform across different stay durations.
- This suggests that length of stay in the city does not significantly influence total spending during the Black Friday event.
- In every stay duration category, there are users with both very low and very high spending.
- There are clear outliers with extremely high purchases in each group, hinting that some users made large bulk purchases, likely businesses or high-net-worth individuals, regardless of how long they lived in the city.
- Customer spending during Black Friday is not dependent on how long they’ve stayed in the city.
- Targeting users solely based on city tenure may not be effective for predicting purchasing behavior in flash sales.

```python
df['Stay_Years_Num'] = df['Stay_In_Current_City_Years'].replace({'4+': 4}).astype(int)

avg_purchase_by_stay = df.groupby('Stay_Years_Num')['Purchase'].mean()

avg_purchase_by_stay.plot(kind='line', marker='o')
plt.title('Average Purchase vs Years in City')
plt.xlabel('Years in Current City')
plt.ylabel('Average Purchase Amount')
plt.grid(True)
```

### Insights:
- The average purchase amount increases steadily from year 0 to year 2. This may suggest that as people settle into a new city, their spending behavior rises up gradually, possibly due to growing comfort or familiarity with local options.
- After peaking at year 2 (~$9395), the average spending starts to decline slightly for users staying 3 and 4 years. This could indicate saturation in buying behavior or shifting priorities after initial setup purchases.
